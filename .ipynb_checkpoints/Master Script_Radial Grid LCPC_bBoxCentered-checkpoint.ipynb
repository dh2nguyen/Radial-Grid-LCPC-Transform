# Part-1

```python
# Part-1 - Notes

# Purpose: This script finds the outer most pixels on four sides (left, top, right, bottom) of the object 
#    and crops the image in a way that leaves a 10-pixel border on each of the four sides.

# Instructions
# 1. The working directory should contain only image files.
# 2. The image files should have a white space of more than 10-pixels surrounding its four sides.  
# 3. For each input image, this script produces an image with the phrase "_resized-crop.png" at the 
#      end of its file name.
```

```python
# Part-1 - Main code

for item in contents:
    
    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Load the image and copy its name as a string for use in renaming output files

    ########
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads an image's name into cv2.imread, which is stored as "the_im",
    # and then passes that name to be returned as a string that is stored as "theName"
    def img_loader(x):
        img = f'cv2.imread("{x}")'
        nombre = name_copier(x)
        return eval(img), nombre
    ########


    # Part 1A.2: Load the image and copy its name
    the_im, theName = img_loader(item)
    plt.imshow(the_im)
    print(theName)

    plt.imshow(cv2.cvtColor(the_im, cv2.COLOR_BGR2RGB))


    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 4 - Crop Image but with a 10-pixel border on 4 sides

    gray = cv2.cvtColor(the_im, cv2.COLOR_BGR2GRAY)
    plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))

    blur = cv2.GaussianBlur(gray, (3,3), 0)
    plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))

    thresh = cv2.threshold(blur, 220, 255, cv2.THRESH_BINARY_INV)[1]
    plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))

    # This finds each indpendent shape in the image as contours. The output ranks the contours from 
    #    smallest to largest, so index 0 is the smallest one: 1st=len(cnts[0][0]), 2nd=len(cnts[0][1]), 3rd=len(cnts[0][2]), etc.
    cnts = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


    # These four lists will collect the x, y, w, or h value from each item in cnts
    list_of_x = []
    list_of_y = []
    list_of_w = []
    list_of_h = []

    # This function appends the x,y,w,h values to the collection lists defined above
    def isolate_contours(input):
        x,y,w,h = cv2.boundingRect(cnts[0][input])
        list_of_x.append(x)
        list_of_y.append(y)
        list_of_w.append(w)
        list_of_h.append(h)

    # This creates an array matching the number of contours in cnts (which is cnts[0])
    numb_of_cnts = np.arange(0,len(cnts[0]))

    # this applies isolate_contours() to each item in numb_of_cnts
    for item in numb_of_cnts:
        isolate_contours(item)

    # list_of_x, list_of_y, list_of_w, list_of_h


    #   xM    xO     yM   yO    wM  wO   hM   hO
    # ([281, 315], [242, 219], [3, 30], [14, 38])


    # Find Left-Most Edge
    # To find the left most edge of all objects, just take the min of list_of_x
    left_edge = min(list_of_x)



    #### Start Section - Find Right-MostEdge

    # To find the right most edge of all objects, you need to first calculate this value for each object. 
    #   Then select the largest value.

    # this list will collect right-most x values for each contour
    list_of_rightside_x = []

    # this the function
    def calc_right_most_x(input):
        rightmost = list_of_x[input] + list_of_w[input]
        list_of_rightside_x.append(rightmost)

    # this is the range over which to iterate calc_right_most_x()    
    numb_of_rightmosts = np.arange(0,len(list_of_x))    

    # this for loop applies iterate calc_right_most_x() across numb_of_rightmosts
    for item in numb_of_rightmosts:
        calc_right_most_x(item)

    # this finds the furthest x within the list    
    right_edge = max(list_of_rightside_x)
    # right_edge   

    #### End Section

    # Find Top-Most Edge    
    # To find the top most edge of all objects, just take the min of list_of_y
    top_edge = min(list_of_y)



    #### Start Section - Find Bottom-Most Edge
    # To find the BOTTOM MOST edge of all objects, you need to first calculate this value for each object. 
    #   Then select the largest value.

    # this list will collect bottom-most y values for each contour
    list_of_bottom_y = []

    # this the function
    def calc_bottom_most_y(input):
        bottom_most = list_of_y[input] + list_of_h[input]
        list_of_bottom_y.append(bottom_most)

    # this is the range over which to iterate calc_bottom_most_y()    
    numb_of_bottom_mosts = np.arange(0,len(list_of_y))    

    # this for loop applies iterate calc_bottom_most_y() across numb_of_bottom_mosts
    for item in numb_of_bottom_mosts:
        calc_bottom_most_y(item)

    # this finds the furthest x within the list    
    bottom_edge = max(list_of_bottom_y)
    # bottom_edge  

    #### End Section


    # left_edge, right_edge, top_edge, bottom_edge    

    # Define new margins -- for cropping -- by adding a 10-pixel border around on four sides
    leftMinus10 = left_edge-10
    rightPlus10 = right_edge+10
    topPlus10 = top_edge-10
    bottomPlus10 = bottom_edge+10

    # leftMinus10, rightPlus10, topPlus10, bottomPlus10

    # Crop the resized image based on the new margins defined in the previous step
    cropped_image = the_im[topPlus10:bottomPlus10, leftMinus10:rightPlus10]
    plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))


    # Save the image by adding a string to the input file's name

    # strip the file extension off of the file name
    split_string = theName.split(".", 1)
    strippedName = split_string[0]

    # add suffix to image that will be saed
    new_name = f'{strippedName}_cropped10.png' 

    # save the stencil image
    cv2.imwrite(new_name, cropped_image) 
```

```python
# Part-2a - Main code 2a


# This empty list will collect results from the for-loop
centers_collector = []

for item in contents:
    
    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Load the image and copy its name as a string for use in renaming output files

    ########
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads an image's name into cv2.imread, which is stored as "the_im",
    # and then passes that name to be returned as a string that is stored as "theName"
    def img_loader(x):
        img = f'cv2.imread("{x}")'
        nombre = name_copier(x)
        return eval(img), nombre
    ########


    # Part 1A.2: Load the image and copy its name
    the_im, theName = img_loader(item)
    # plt.imshow(the_im)
    # print(theName)

    #plt.imshow(cv2.cvtColor(the_im, cv2.COLOR_BGR2RGB))


    # Convert the image to gray scale
    gray = cv2.cvtColor(the_im, cv2.COLOR_BGR2GRAY)
    #plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))


    # invert the image so that the object is white
    inverted = cv2.bitwise_not(gray)

    #plt.imshow(cv2.cvtColor(inverted, cv2.COLOR_BGR2RGB))

    # Do Canny edge detection to find outlines of the object

    # Setting parameter values for edge detection
    t_lower = 50  # Lower Threshold
    t_upper = 150  # Upper threshold

    # Applying the Canny Edge filter
    edge = cv2.Canny(inverted, t_lower, t_upper)
    #plt.imshow(cv2.cvtColor(edge, cv2.COLOR_BGR2RGB))

    # cv2.findContours detects the contours of the edges detected by Canny edge detection
    contours, hierarchy = cv2.findContours(edge, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    # contours

    list_of_lengths = [] # this list collects the length of each object in "contours"

    for item in np.arange(0,len(contours)):
        how_long = len(contours[item])
        list_of_lengths.append(how_long) 

    longest_contour = max(list_of_lengths) # this selects the longest length in list_of_lengths
    #longest_contour

    list_of_tuples = [] # this list collects the length of each object in "contours"

    for idx, item in enumerate(contours):
        how_long = len(contours[idx])
        row_num = idx
        tuple_of_interest = (row_num, how_long)
        list_of_tuples.append(tuple_of_interest) 

    list_of_tuples = pd.DataFrame(list_of_tuples, columns =['RowNum', 'Length']) # turn list_of_tuples into a dataframe
    #list_of_tuples

    # get the row index of the object in contours that has the longest length
    index_of_largest_cont = list_of_tuples['Length'].idxmax()
    #index_of_largest_cont

    # this creates a white canvas of the same size as the input image
    mask = np.ones(the_im.shape[:2], dtype="uint8") * 255

    # this draws the contours -- in BLUE color -- on a white canvas that is the same dimensions as the input image
    outline = cv2.drawContours(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB), contours, -1, (0,0,255), 1)
    #plt.imshow(outline)


    # Get the coordinates of the center of the bounding box of the image
    x, y, w, h = cv2.boundingRect(gray) # Find upperleft xy coordinate of image, width of image, and height of image
    cx = x + (w // 2) # Find midpoint of width. Assign as cx
    cy = y + (h // 2) # Find midpoint of height. Assign as cy

    theta_x_y = (theName, cx, cy) # this collects as a tuple: theta, cX, and cY
    centers_collector.append(theta_x_y) # Append image name, cx, and cy as a tuple


    # Calculate the centroid of the object 
    #M = cv2.moments(contours[index_of_largest_cont])

    #cX = int(M["m10"] / M["m00"])
    #cY = int(M["m01"] / M["m00"])
    #cX, cY

    #triple_binder = tuple((theName, cX, cY))
    #centroid_collector.append(triple_binder)
    #cX = []
    #cY = []

    
centers_collector


# Turn the list of tuples into a dataframe for saving as a csv file
results = pd.DataFrame(centers_collector, columns =['name', 'x_coord', 'y_coord'])
#results

# Define generic name for the output csv file
file_name = 'names_and_centers.csv' 

# save the dataframe as a csv file
results.to_csv(file_name, index=False)


```

```python
# Part-3 - Main code


######################################################################
######################################################################
# These are the functions required for the main code that iterates across images in the working directory.

# The following functions take in the starting point of a line, it's length, it's angle, and 
#   then calculates the endpoint of the line.

# For each quadrant in the Cartesian plane, you need to calculate the end point of the line that starts at the origin
#   and that extends at a specific angle within that quandrant for a length that is stored in "half_diagonal". 
# I used the SOH-CAH-TOA Trigonometric equations. These only work for angles that are not 0, 90, 180, 270, and 360. 


def get_xy_of_angle_0to90_notinclusive(theta):
    angle = theta
    adjacent = half_diagonal*(math.cos(theta*(np.pi/180))) # this calculates the length of the side adjacent to theta (derived from cos(theta)=adjacent/hypotenuse)
    x_end = int(x_origin + adjacent) # the x-coordinate of the endpoint the adjacent side is (x_origin + adjacent)
    
    opposite = half_diagonal*(math.sin(theta*(np.pi/180))) # this calculates the length of the side opposite to theta (derived from sin(theta)=opposite/hypotenuse)
    y_end = int(y_origin - opposite) # the y-coordinate of the endpoint the opposite side is (y_origin - opposite because y=0 is the top of the image) 
    
    xy_of_endpoints.append(tuple((angle, x_end, y_end))) # this appends theta and its corresponding x_end and y_end to results_collector
    

def get_xy_of_angle_90to180_notinclusive(theta):
    angle = theta 
    adjacent = half_diagonal*(-math.cos(angle*(np.pi/180))) # in quadrant II, cosines are negative, so need to reverse the negative
    x_end = int((x_origin - adjacent)) 
    
    opposite = half_diagonal*(math.sin(angle*(np.pi/180))) # this calculates the length of the side opposite to theta (derived from sin(theta)=opposite/hypotenuse)
    y_end = int(-(opposite - y_origin)) # remember that y=0 is the top left of the image 
    
    xy_of_endpoints.append(tuple((angle, x_end, y_end))) 
    
    
def get_xy_of_angle_180to270_notinclusive(theta):
    angle = theta 
    adjacent = half_diagonal*(-math.cos(angle*(np.pi/180))) # in quadrant III, cosine is negative, so need to reverse the negative
    x_end = int((x_origin - adjacent)) 
    
    opposite = half_diagonal*(-math.sin(angle*(np.pi/180))) # in quadrant III, sin is negative, so need to reverse the negative
    y_end = int((opposite + y_origin)) # remember that y=0 is the top left of the image 
    
    xy_of_endpoints.append(tuple((angle, x_end, y_end))) 

    
def get_xy_of_angle_270to360_notinclusive(theta):
    angle = theta 
    adjacent = half_diagonal*(math.cos(angle*(np.pi/180))) # in quadrant III, cosine is negative, so need to reverse the negative
    x_end = int((x_origin + adjacent)) 
    
    opposite = half_diagonal*(-math.sin(angle*(np.pi/180))) # in quadrant III, sin is negative, so need to reverse the negative
    y_end = int((opposite + y_origin)) # remember that y=0 is the top left of the image 
    
    xy_of_endpoints.append(tuple((angle, x_end, y_end))) 

    
    
# For angles that are 0, 90, 180, 270, or 360, the following are functions calculate the end point of lines that fall on them.

def get_xy_of_angle_0_90_180_270_360(theta):
    angle = theta
    if angle == 0:
        x_end = 2 * x_origin
        y_end = y_origin
        xy_of_endpoints.append(tuple((angle, x_end, y_end))) 
    elif angle == 90:
        x_end = x_origin
        y_end = 0
        xy_of_endpoints.append(tuple((angle, x_end, y_end))) 
    elif angle == 180:
        x_end = 0
        y_end = y_origin
        xy_of_endpoints.append(tuple((angle, x_end, y_end))) 
    elif angle == 270:
        x_end = x_origin
        y_end = 2 * y_origin
        xy_of_endpoints.append(tuple((angle, x_end, y_end))) 
    elif angle == 360:
        x_end = 2 * x_origin
        y_end = y_origin
        xy_of_endpoints.append(tuple((angle, x_end, y_end))) 


######################################################################
######################################################################

# This function takes in an angle and decides which endpoint-calculating function from above to call based on which 
#   Cartesian quadrant the angle belongs to.

def endpoint_finder(input):
    angle = input
    if angle > 0 & angle < 90:
        get_xy_of_angle_0to90_notinclusive(angle)
    elif angle > 90 & angle < 180:
        get_xy_of_angle_90to180_notinclusive(angle)
    elif angle > 180 & angle < 270:
        get_xy_of_angle_180to270_notinclusive(angle)
    elif angle > 270 & angle < 360:
        get_xy_of_angle_270to360_notinclusive(angle)
    elif angle == 0:
        get_xy_of_angle_0_90_180_270_360(angle)
    elif angle == 90:
        get_xy_of_angle_0_90_180_270_360(angle)
    elif angle == 180:
        get_xy_of_angle_0_90_180_270_360(angle)
    elif angle == 270:
        get_xy_of_angle_0_90_180_270_360(angle)
    elif angle == 360:
        get_xy_of_angle_0_90_180_270_360(angle)
    

############################################################################################################################################
############################################################################################################################################
############################################################################################################################################



# This function calls each key:value pair and then calculates centroids
def intersection_centroid(input):
    angle_as_int = int(input)
    bundle = dict_of_contours.get(input)
    for thing in np.arange(0,len(bundle)):
        M = cv2.moments(bundle[thing]) 

        # We need this if-else statement to ignore objects that are open shapes, meaning 
            #   they are so thin that Python does not think they are closed shapes. If Python thinks that 
            #   something is not a closed shape, it returns zeros for all the moments. Dividing by zero triggers an error. 
        if M["m00"] == 0:
            pass
        else: 
            cX = int(M["m10"] / M["m00"]) # 8. get centroid of object
            cY = int(M["m01"] / M["m00"])
            theta_x_y = (angle_as_int, cX, cY) # this collects as a tuple: theta, cX, and cY.
            angles_and_centroids.append(theta_x_y) # 9. record centroid coordinates and angle input into the list called "angles_and_centroids"



######################################################################
######################################################################

# Part-3 - Main code

xy_of_endpoints = [] # this empty list will collect xy coordinates the endpoint-calculating functions

for item in image_list:     
    
    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Load the image and copy its name as a string for use in renaming output files

    ########
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads an image's name into cv2.imread, which is stored as "the_im",
    # and then passes that name to be returned as a string that is stored as "theName"
    def img_loader(x):
        img = f'cv2.imread("{x}")'
        nombre = name_copier(x)
        return eval(img), nombre
    ########


    # Part 1A.2: Load the image and copy its name
    the_im, theName = img_loader(item)
    #plt.imshow(the_im)
    #print(theName)

    x_origin, y_origin = xy_matcher(theName)

    plt.imshow(cv2.cvtColor(the_im, cv2.COLOR_BGR2RGB))



    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 3 - Dilate the image 


    # Convert the image to gray scale
    gray = cv2.cvtColor(the_im, cv2.COLOR_BGR2GRAY)
    # plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))

    # invert the image so that the object is white
    inverted = cv2.bitwise_not(gray)

    plt.imshow(cv2.cvtColor(inverted, cv2.COLOR_BGR2RGB))


    # Define the matrix kernel dimensions by which to dilate the object
    kernel = np.ones((3,3), np.uint8)

    # This does the dilation
    dilated = cv2.dilate(inverted, kernel, iterations=1)

    plt.imshow(cv2.cvtColor(dilated, cv2.COLOR_BGR2RGB))



    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 4a - Define the origin
    # Step 4b - Define the desired spacing between the radial lines 


    ####################################################################################
    # Step 4a

    # define the x and y coordinates of the origin
    # x_origin is defined in the function called xy_matcher()
    # y_origin is defined in the function called xy_matcher()


    # define width and height of image
    img_width = the_im.shape[1]
    img_height = the_im.shape[0]
    # img_width, img_height


    # calculate the diagonal of length of the image (pythagorean theorem)
    diagonal = math.sqrt( (img_width**2) + (img_height**2) )

    # define half the length of the image's diagonal as the length of each line of the radial grid
    # half_diagonal will be used as the length of the hypotenuse for the following trigonometric functions
    half_diagonal = int(diagonal/2)

    # diagonal, half_diagonal


    ####################################################################################
    # Step 4b

    # Define how many degrees between each radial line from 0 to 360
    spacing_bt_lines = 5

    # Calculate the number of lines based on the desired spacing between lines
    number_of_lines = 360/spacing_bt_lines
    number_of_lines

    # create list of angles according to the desired spacing between radial lines
    angles = list(range(0,360, spacing_bt_lines))
    # angles



    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 5 - Draw a line, create a stencil, overlay stencil on original, then extract exposed region

    # Need to draw the line for each desired angle one at a time. The cycle consists of the following steps. 
    # 1. take angle input
    # 2. calculate endpoints
    # 3. draw the line
    # 4. create stencil
    # 5. overlay stencil on orginal image
    # 6. threshold for blue color
    # 7. get contour of blue object
    # 8. get centroid of object
    # 9. record centroid coordinates and angle input


    # This function calculates the centroid for a contour, then appends the x&y coordinates -- along with
        #    the value of "theta" -- to the list called "angles_and_centroids".

    def centroid_finder(input):
        M = cv2.moments(contours[input]) 

        # We need this if-else statement to ignore objects that are open shapes, meaning 
            #   they are so thin that Python does not think they are closed shapes. If Python thinks that 
            #   something is not a closed shape, it returns zeros for all the moments. Dividing by zero triggers an error. 
        if M["m00"] == 0:
            pass
        else: 
            cX = int(M["m10"] / M["m00"]) # 8. get centroid of object
            cY = int(M["m01"] / M["m00"])

        theta_x_y = (theta, cX, cY) # this collects as a tuple: theta, cX, and cY
        angles_and_centroids.append(theta_x_y) # 9. record centroid coordinates and angle input into the list called "angles_and_centroids"


    dict_of_contours = {} # this empty dictionary will store each angle and its corresponding contour objects 


    for item in angles:
        # 1. take angle input    
        theta = item

        # 2. calculate endpoints
        endpoint_finder(theta)

        # xy_of_endpoints 

        # 3. draw the line
        cv2.line(dilated, (x_origin,y_origin), (xy_of_endpoints[0][1:3]), color = (0,0,0), thickness = 3 )
        # plt.imshow(cv2.cvtColor(dilated, cv2.COLOR_BGR2RGB))


        xy_of_endpoints = [] # this clears the contents of this list so that it can be replaced by info from the next iteration


        # 4. Create stencil: Find contours of mask and redraw contours on original image to reveal intersections.
        # This turns the image into a binary black and white image (objects should be white, background should be black)
        ret, binary = cv2.threshold(dilated, 20, 255, cv2.THRESH_OTSU)
        # plt.imshow(binary, cmap="gray")


        # cv2.findContours detects the contours of the white shapes in the image
        contours, hierarchy = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


        # this step is necessary to reset the dilated image to its original form so that new intersections are placed on
        #    the original dilated image
        dilated = cv2.dilate(inverted, kernel, iterations=1)


        # 5. overlay stencil on orginal image
        # this draws the contours -- in BLACK color -- on the original input image
        stencil_overlay = cv2.drawContours(cv2.cvtColor(the_im, cv2.COLOR_BGR2RGB), contours, -1, (0,0,0), thickness=cv2.FILLED)
        # plt.imshow(stencil_overlay)


        # 6. threshold for blue color: Select only blue pixels and create a mask 
        hsv = cv2.cvtColor(stencil_overlay, cv2.COLOR_RGB2HSV)
        lower_range = np.array([110,50,50])
        upper_range = np.array([130,255,255])
        intersections = cv2.inRange(hsv, lower_range, upper_range)

        # plt.imshow(intersections, cmap="gray")


        # 7. get contour of blue objects
        # this turns the image into a binary black and white image (objects should be white, background should be black)
        ret, binary = cv2.threshold(intersections, 20, 255, cv2.THRESH_OTSU)
        # plt.imshow(binary, cmap="gray")

        # this detects the contours of the white shapes in the image
        contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Stored each contour's set of objects as the values of a dictionary key: key = theta, value = contours set
        dict_of_contours[f'{theta}'] = contours


    # this is an empty list that will collect results from the below for-loop
    angles_and_centroids = []

    # dictinoary keys must be turned into a list of strings before each key can be indexed by a number
    keys_as_list = list(dict_of_contours) 

    # iterate centroid_calc() across keys_as_list
    for stuff in keys_as_list:
        intersection_centroid(stuff)
        
    angles_and_centroids


    x_origin = [] # this clears the value of this variable in preparation for the next iteration, to avoid confusion
    y_origin = [] # this clears the value of this variable in preparation for the next iteration, to avoid confusion

    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 6 - Save the results as a dataframe and then export as csv file


    # 10. Turn the list of tuples into a dataframe for saving as a csv file
    results = pd.DataFrame(angles_and_centroids, columns =['angle', 'x_coord', 'y_coord'])

    # save result as a csv file
    # strip the file extension off of the file name
    split_string = theName.split(".", 1)
    strippedName = split_string[0]

    # add suffix to image that will be saved
    new_name = f'{strippedName}_intersections.csv' 

    # save the dataframe image
    results.to_csv(new_name, index=False)
    

```

```python
# Part-5 - Main code


for item in intersection_files:
    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 1 - Load the dataframe and copy its name as a string for use in renaming output files

    ########
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads an dataframe's name into pd.read_csv, which is stored as "the_df",
    # and then passes that name to be returned as a string that is stored as "theName"
    def df_loader(x):
        df = f'pd.read_csv("{x}")'
        nombre = name_copier(x)
        return eval(df), nombre
    ########


    # Part 1A.2: Load the dataframe and copy its name
    the_df, theName = df_loader(item)
    # theName

    # the intersection file's name needs to be stripped  of "_intersections.csv" and appended with ".png"

    # strip the file extension off of the file name
    split_string = theName.split("_intersections", 1)
    strippedName = split_string[0]

    # add suffix to image that will be saved
    new_name = f'{strippedName}.png' 
    new_name


    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 2 - Calculate Distances


    # this searches for the x and y coordinates in dataframe called "tracker"
    x_origin, y_origin = xy_matcher(new_name)
    # x_origin, y_origin

    # Find number of rows in the dataframe
    numRows = the_df.shape[0]
    # numRows

    # Create a list repeating x_origin the same number of times as the number of rows in the dataframe
    the_x = [x_origin]*numRows
    the_y = [y_origin]*numRows

    # add the_x and the_y as new columns in the_df
    the_df['origin_x'] = the_x
    the_df['origin_y'] = the_y
    # the_df

    # This unction calculates distance between two points
    #     Distance Equation between two points in Cartesian space:
    #     sqrt( (x-x0)**2 + (y-y0)**2 )
    def dist_calc(input):
        x1 = the_df.iloc[input, 1]
        x2 = the_df.iloc[input, 3]
        y1 = the_df.iloc[input, 2]
        y2 = the_df.iloc[input, 4]
        the_dist = int(math.sqrt( (x2-x1)**2 + (y2-y1)**2 ))
        distance.append(the_dist)

    distance = [] # this empty list will collect distance values

    # this for-loop applies the dist_calc() function across the row index of the_df
    for item in np.arange(0,numRows):
        dist_calc(item)

    # add the list called "distance" as a column in the_df
    the_df['distance'] = distance
    the_df


    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 3 - Sum the Distances for angles that have more than one intersection

    # get unique elements in the column called "angle"
    unique_angles = the_df.angle.unique()
    # unique_angles

    # this empty list will collect summed distances for each angle
    lcpc = []

    # This function sums distances in all rows of the column called "distance"
    def dist_summer(input):
        angle = input
        sub_df = the_df.loc[the_df['angle'] == input]
        summed_dist = sub_df['distance'].sum()
        lcpc.append((angle,summed_dist))

    for item in unique_angles:
        dist_summer(item)

    # lcpc

    ####################################################################################
    ####################################################################################
    ####################################################################################
    # Step 4 -  Save the results as a dataframe and then export as csv file


    # 10. Turn the list of tuples into a dataframe for saving as a csv file
    results = pd.DataFrame(lcpc, columns =['angle', 'summed_distances'])

    # save result as a csv file
    # strip the file extension off of the file name
    split_string = theName.split(".", 1)
    strippedName = split_string[0]

    # add suffix to image that will be saved
    new_name = f'{strippedName}_LCPC-data.csv' 

    # save the dataframe image
    results.to_csv(new_name, index=False)


```

# Part-6

```python
# Part-6 - Notes


# This script does the FFT on a column of data. 
#   This script outputs a csv file with the words "_fft-output_AllBins" at the end of its file name.

# NOTE 1: The sampling rate is set to 40. See variabled called "samp_rate".
# We pick a sampling rate of 40 because the original way we did it was to divide a 200 pixel-tall image by 5 pixels, 
#  resulting in 40 horizontal lines. Changing the sampling rate changes the resulting frequencies
#  but does NOT change the resulting magnitudes. Since we are interested in just the magnitudes and not the
#  frequencies, the sampling rate doesn't really matter. 

    
# How To Use This Script:    
#   This script takes a csv file that was outputted from the previous step. 
#   The input file must have "LCPC-data.csv" at the end of its file name. Each unique sample should only have one 
#      file that has this phrase in it's file name. 
#   The input file must have a column called "summed_distances".
```

```python
# Call working directory again to refresh the list of files in it

os.chdir(working_directory)
path = os.getcwd()
#print(path)
contents = os.listdir(path)


# Create a list of files that end with "_cropped10_intersections_LCPC-data.csv" in their name
lcpc_files = []
for item in contents:
    if item.endswith("_cropped10_intersections_LCPC-data.csv") == True:
        lcpc_files.append(item)

lcpc_files   
```

```python
# Part-6 - Main code

for item in lcpc_files:
    
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads a file name into pd.read_csv, which is stored as "table",
    # and then passes that file to the name_copier function to get a string that is stored as "nombre"
    def df_reader(x):
        table = f'pd.read_csv("{x}")'
        nombre = name_copier(x)
        return eval(table), nombre
    ########

    df, df_name = df_reader(item)
    # df


    # turn the column calld 'summed_distances' into a list
    data = df['summed_distances'].tolist()
    # data



    # Define the sampling rate. See the introduction of this script for why a sampling rate of 40 was chosen. 
    samp_rate = 40

    # Do the FFT
    fft_result = fft(data)
    N = len(fft_result)
    n = np.arange(N)
    T = N/samp_rate
    freq = n/T 

    # ####
    # # The next few lines of code are for plotting the fft spectrum
    # plt.figure(figsize = (12, 6))
    # plt.subplot(121)

    # plt.stem(freq, np.abs(X), 'b', markerfmt=" ", basefmt="-b")
    # plt.xlabel('Freq (Hz)')
    # plt.ylabel('FFT Amplitude |X(freq)|')
    # plt.xlim(0, 20)
    # plt.show()
    # ####


    ###############################
    # calculate the magnitudes by taking absolute value of the variable called fft_result
    magnitudes = np.abs(fft_result)

    # calculate the single-sided magnitude by multiplying the magnitudes by 2
    single_sided_magnitudes = magnitudes * 2


    # create a list of numbers to track each row along the length of fft_result, which is the variable N
    counter = np.arange(1,N,1)

    ################################
    # combine counter, freq, and single_sided_magnitudes into a dataframe
    # This script does collects ALL frequencies for each sample, not just the smallest 19.    
    final_df = pd.DataFrame(list(zip(counter, freq, single_sided_magnitudes)), columns=['order','freq','single.sided.magnitude'])
    # final_df


    #################################
    # save result as a csv file

    # strip the file extension off of the file name
    split_string = df_name.split("LCPC-data", 1)
    strippedName = split_string[0]

    # add suffix to image that will be saved
    new_name = f'{strippedName}LCPC-data_fft-output_AllBins.csv' 

    # save the dataframe image
    final_df.to_csv(new_name, index=False)


```

```python
# Part-7 - Notes


# Purpose: This script isolates the magnitude column from each fft ouput from Step 4 and combines them all 
#  into one dataframe.


# How To Format the Input File:
#   Each input file should have the phrase "fft-output" in it's file name. Each sample should have only ONE csv file
#     that has this phrase in it.
#   Each input file should have a column whos header is called "single.sided.magnitude". 
#   Each input file should have the word "mask" in it's name or the naming of the columns in the output file will 
#     have a very long name, which still works. 

#   This script creates a csv output file with the generic name of "compiled_fft_RadialGridLCPC.csv"
```

```python
# Call working directory again to refresh the list of files in it

os.chdir(working_directory)
path = os.getcwd()
#print(path)
contents = os.listdir(path)

# Create a list of files that have the word "_cropped10_intersections_LCPC-data_fft-output_AllBins" in their name
substring = '_cropped10_intersections_LCPC-data_fft-output_AllBins'
contents = [i for i in contents if substring in i]
sorted_contents = sorted(contents) # sort alphabetically from a to z
sorted_contents
```

```python
# create an empty list that will receive contents later
list_o_numRows = []

# This for-loop opens a csv file, counts the number of rows in it, and then appends that number to list_o_numRows
for item in sorted_contents:
    df = pd.read_csv(item)
    how_many = df.shape[0]
    list_o_numRows.append(how_many) 

# Find the largest number in list_o_numRows, which will define the height of the empty dataframe that comes next.
#   This is necessary or the below for-loop will cut off rows from certain csv files
df_max_height = max(list_o_numRows)

# list_o_numRows, df_max_height


# Create an empty dataframe that will receive content later that has the number of rows specified by 
#   df_max_height and taking into account Python's indexing from 0
collector = np.arange(1,(df_max_height+1),1)
collector = pd.DataFrame(collector,columns=['bin_order'])
# collector


for item in sorted_contents:

    ########
    # this function copies the name of the file that is passed into and returns that name
    def name_copier(y):
        fileName = y
        return(fileName)

    # This function returns TWO values. This function loads a file name into pd.read_csv, which is stored as "table",
    # and then passes that file to the name_copier function to get a string that is stored as "nombre"
    def df_reader(x):
        table = f'pd.read_csv("{x}")'
        nombre = name_copier(x)
        return eval(table), nombre
    ########

    # the name of the input file is stored in the variable called df. it's name is stored as df_name
    df, df_name = df_reader(item)
    # df

    # isolate the column called single.sided.magnitude
    mags = df["single.sided.magnitude"]
    # mags

    # Split the name of the input file at the word "mask" and copy everything that comes before the word "mask"
    split_string = df_name.split("mask", 1)
    strippedName = split_string[0]

    # rename the column header with the modified file name stored in strippedName
    collector[strippedName] = mags
    #collector


# save collector as a csv file with a generic name
collector.to_csv("compiled_fft_RadialGridLCPC.csv", index=False)
```
